in addition to floats and integers, I want to introduce a new data type called a **category**, and this is something that's unique to pandas. It doesn't exist in Python. 

So a category is a data type that's really ideal when you have a small number of unique values in a dataframe. So if you can imagine a dataframe of, let's say, employees, and it has a million rows, and there's a gender column where you only have two possible values, male or female, or maybe you have a million rows and you have a month column, and it only has 12 possible values, 12 different unique things, but they're all duplicated throughout those million rows. 

Whenever you have a small number of unique values, the category data type can help you severely reduce the memory usage. 

```python
nba['Position'].nunique()
```

In the above there are 5 unique positions in nba dataset

 So we have 5 unique values within the position column out of a total of something like 450 rows. That's a very small percentage. 
 
 So that's the ideal column that we would want to turn into a category. So we don't have to store 450 different unique strings.
 
  Instead, we're only going to have 5 objects, one for each type of position. And every time we run into identical values like PG or SG, it's just going to point to that same thing in memory

```python
nba['Position'].astype('category')
```

```python
nba['Position'] = nba['Position'].astype('category')
nba.info()
```
