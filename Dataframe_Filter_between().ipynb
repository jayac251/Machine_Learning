```python
import pandas as pd

# mounting google drive
from google.colab import drive
drive.mount("/gdrive")

df = pd.read_csv("/gdrive/My Drive/Colab Notebooks/employees.csv")
df.head()
```

#**Data Cleansing**


```python
#Convert Start date, Last login time to date time object

df['Start Date'] = pd.to_datetime(df['Start Date'])

df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])




# Convert Sr.mgmt to Boolean object

df['Senior Management'] = df['Senior Management'].astype('bool')




#Convert Gender to category object
df['Gender'] = df['Gender'].astype('category')
```

#1.pull out all of the employees who had a salary between 60,000 and 70,000 inclusive

```python
filter = df['Salary'].between(60000,70000)

df[filter].sort_values('Salary')
```

#2.pull the rows or those employees that have a bonus between 2 and 5%

```python
filter = df['Bonus %'].between(2,5)

df[filter].sort_values('Bonus %')
```

#3.pull out all of the employees who started in my year of birth, which is 1991

```python
filter = df['Start Date'].between('1991-01-01','1992-01-01')

df[filter].sort_values('Start Date')
```

#4. find the people who logged into the system between 8:30 AM and noon, which is 12 PM

```python
filter = df['Last Login Time'].between('2021-10-12 08:30AM','2021-10-12 12:00PM')

#or

filter = df['Last Login Time'].between('08:30AM','12:00PM')

df[filter].sort_values('Last Login Time')
```
