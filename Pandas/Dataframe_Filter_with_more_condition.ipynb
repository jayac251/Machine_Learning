```python
import pandas as pd

# mounting google drive
from google.colab import drive
drive.mount("/gdrive")

df = pd.read_csv("/gdrive/My Drive/Colab Notebooks/employees.csv")
df.head()
```

#**Data Cleansing**



```python
#Convert Start date, Last login time to date time object

df['Start Date'] = pd.to_datetime(df['Start Date'])

df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])




# Convert Sr.mgmt to Boolean object

df['Senior Management'] = df['Senior Management'].astype('bool')




#Convert Gender to category object
df['Gender'] = df['Gender'].astype('category')
```

# **Filter based on multiple conditions**

##1. extract the rows from the DataFrame where the value in the Gender column is equal to Male, and the value in the Team column is equal to Marketing

```python
c_Gender = df['Gender'] == 'Male'
c_Team = df['Team'] == 'Marketing'

df[c_Gender & c_Team]

```

##2.pull rows that fit one of 2 conditions, or perhaps both. Either a value of true in the Senior Management column or a Start Date before January 1st,1990

```python
c_Srmgmt = df['Senior Management']
c_StrtDate = df['Start Date'] <= '1990-01-01'

df[c_Srmgmt | c_StrtDate]
```
