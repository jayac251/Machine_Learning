# <u>Time Series Analysis</u>

# Data Preparation

```python
from datetime import date
from nsepy import get_history
import pandas as pd
```

```python
symbols = "NIFTY 500,NIFTY 100,NIFTY 200,NIFTY NEXT 50".split(',')
symbols

```

```python
#nifty_next50 = get_history(symbol=stocks,
 #                           start=date(2021,1,1),
 #                           end=date(2021,10,31),
 #                           index=True) 

data={}
for s in symbols:
    data[s] = get_history(symbol=s,
                           start=date(2019,1,1),
                           end=date(2021,10,31),
                           index=True)['Close']
 
```

```python
data
```

```python
df_comp= pd.DataFrame(data)
df_comp
```

```python
df_comp.describe()
```

```python
df_comp.isna().sum()
```

# Visualization

```python
import matplotlib.pyplot as plt
```

```python
df_comp['NIFTY 100'].plot(title = "NIFTY 100")
df_comp['NIFTY 500'].plot(title = "NIFTY 500")
df_comp['NIFTY 200'].plot(title = "NIFTY 200")
df_comp['NIFTY NEXT 50'].plot(title = "NIFTY NEXT 50")
plt.legend()
plt.show()

```


# <u>QQPlot (Quantile -Quantile Plot)</u>

### QQplot takes all the values a variable can take and arranges them in ascending order

<b>Y- axis</b>  --> Price with the highest ones at the top and the lowest at the bottom

<b>X- axis</b>  -->  Represents the theoretical quintiles of the dataset (How many standard deviations away from the mean these values are)

The <font color='red'> <b>red</b> </font> diagonal line on the screen represents what the data points should follow if they are normally distributed.

In this case, the data is normally distributed to some extent

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_comp['NIFTY 100'], plot = pylab)
pylab.show()
```

```python
scipy.stats.probplot(df_comp['NIFTY 500'], plot = pylab)
pylab.show()
```
